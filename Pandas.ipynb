```python
Dealing with mixed data technique:

1. Dropping missing values:
one way to deal with missing data is to drop it.
This is particularly useful in cases where you have plenty of data and losing a small portion won't impact the downstream analyisis.

you can use a .dropna() method

df1 = df.dropna()
```

```python
The axis argument lets you specify whether you are dropping rows, or columns with missing values. The default axis removes the rows containing NaNs.
use axis = 1 to remove the columns with one or more NaN values. Also, notice how we are using the argument inplace = True which lets you skip saving 
the output of .dropna() into a new DataFrame.
```

```python
import pandas as pd
df = pd.read_csv(r"F:/python projects/food.csv")
df.head()

#it will display the first 5 rows of the data.
```

```python
df.dropna(inplace =True, axis = 0)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
#2) Replacing the missing value:
import pandas as pd
df = pd.read_csv("F:/python projects/food.csv", usecols = ['thai_food', 'vitamins', 'turkey_calories'])
df.tail(50)
```

```python
#replacing the NaN value usinf mean()

mean_value = df['vitamins'].mean()
print(mean_value)
df1 = df.fillna(mean_value)
df1.tail(20)
```
