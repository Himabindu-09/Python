```python
#Matplotlib Plotting:

Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.
Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.
```

```python
#example: Draw a line in a diagram from position (1, 3) to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

x = np.array([6, 4])
y = np.array([9, 12])

plt.plot(x, y)
plt.show()
```

```python
Types of Matplotlibs:

Matplotlib comes with a wide variety of plots. plots help to understand trends. and patterns, and to make correlations.
They're typically instruments for reasoning about qunatitative information. Some of the plots are covered here.

*Matplotlib Line plot
*Matplotlib Bar plot
*Matplotlib Histogram plot
*Matplotlib Scatter plot
*Matplotlib Pie plot
*Matplotlib Area plot
```

```python
#Matplotlib line plot:

#importing matplotlib module.

import matplotlib.pyplot as plt
#x-axis values:
x = [5,7,3,5,3,9,7]

#y-axis values:
y = [8,3,6,2,1,9,5]
#function to plot
plt.plot(x,y)
#function to show the plot
plt.show()
```

```python
#Matplotlib Bar plot:

#importing matplotlib module.

import matplotlib.pyplot as plt
#x-axis values:
x = [5,7,3,5,3,9,7]

#y-axis values:
y = [8,3,6,2,1,9,5]
#function to plot
plt.bar(x,y)
#function to show the plot
plt.show()
```

```python
#Matplotlib Histogram plot:

#importing matplotlib module.

import matplotlib.pyplot as plt

#y-axis values:
y = [8,3,6,2,1,9,5]
#function to plot
plt.hist(y)
#function to show the plot
plt.show()
```

```python
#Matplotlib Scatter plot:

#importing matplotlib module.

import matplotlib.pyplot as plt
#x-axis values:
x = [5,7,3,5,3,9,7,8,3,1]

#y-axis values:
y = [8,3,6,2,1,9,5,9,1,6]
#function to plot
plt.scatter(x,y)
#function to show the plot
plt.show()
```

```python
#Matplotlib Pie plot:

import matplotlib.pyplot as plt
#Data for the pie chart
labels = ['Apples', 'bananas', 'oranges']
size = [56,23,26]

#plotting the pie chart

plt.pie(size, labels = labels, autopct = '%1.1f%%', startangle =90)
plt.title("Pie Chart")
plt.show()
```

```python
#Matplotlib Area plot:

import matplotlib.pyplot as plt
#Data

x = [3,6,5,2,7]
y1,y2 = [45,78,12,54,23],[7,14,21,28,35]
#Area Chart
plt.fill_between(x,y1,y2,color = 'purple', alpha = 0.4, label = "Area covered")
plt.plot(x,y1,label = "line 1", marker = "*")
plt.plot(x,y2,label = "line 2", marker = "*")

#labels and title
plt.xlabel("x-axis"),plt.ylabel("y-axis"),plt.title("Area Chart")

#Legend and display
plt.legend(),plt.show()
```

```python
#introduction to figures:

#constrained layout automatically adjusts subplots.

fig =plt.figure(figsize =(2,3), facecolor = "blue", layout = "constrained")
fig.suptitle("Figure")
ax = fig.add_subplot()
ax.set_title("Axes", loc = "Right", fontstyle = "italic", fontsize = "xx-large")
```

```python
#subplot() function:

#plt.subplot(subplot(nrows, ncols, index))
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot([1,2,3,4,5,6])
ax2 = fig.add_subplot(221, facecolor = 'y')
ax2.plot([1,2,3,7,4,2])
```

```python
fig,a = plt.subplots(2,2)
import numpy as np
x = np.arange(1,6)
a[0][0].plot(x, x*x)
a[0][0].set_title("SQUARE")
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title("SQUARE ROOT")
a[1][0].plot(x, np.exp(x))
a[1][0].set_title("EXPONENTIAL")
a[1][1].plot(x, np.log10(x))
a[1][1].set_title("LOG")
plt.show()

```
